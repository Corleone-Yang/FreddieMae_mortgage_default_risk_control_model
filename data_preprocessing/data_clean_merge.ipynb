```python
def insertNull(filePath):
    with open(filePath, 'r') as file:
        lines = file.readlines()
    processedData = []
    for line in lines:
        fields = line.strip().split("|")
        fields = [item if item != '' else None for item in fields]
        processedData.append(fields)
    return processedData
```

```python
import pandas as pd
def convert_orig(filepath,outputpath):
    processedData=insertNull(filepath)
    ODF_df=pd.DataFrame(processedData)
    ODF_df.columns = ['CS',#CREDIT SCORE
              'FPD', #FIRST PAYMENT DATE
              'FTHF', #FIRST TIME HOMEBUYER FLAG
              'MD', #MATURITY DATE
              'MSA/MD', #METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION -
              'MIP', #MORTGAGE INSURANCE PERCENTAGE
              'NU', #NUMBER OF UNITS
              'OS', #OCCUPANCY STATUS
              'OCLTV', #original combined loan-to-value
              'ODTIR', #ORIGINAL DEBT-TO-INCOME (DTI) RATIO
              'OU', #ORIGINAL UPB
              'OLTV', #ORIGINAL LOAN-TO-VALUE
              'OIR', #ORIGINAL INTEREST RATE
              'CHANNEL',
              'PPMF', #PREPAYMENT PENALTY MORTGAGE (PPM) FLAG
              'AT', #AMORTIZATION TYPE
              'PS', #PROPERTY STATE
              'PT', #PROPERTY TYPE
              'PC', #POSTAL CODE
              'LSN', #LOAN SEQUENCE NUMBER
              'LP', #LOAN PURPOSE
              'OLT', #ORIGINAL LOAN TERM
              'NB', #Number of borrowers
              'SellerName',#SELLER NAME
              'ServicerName', #ServicerName
              'SCF', #SUPER CONFORMING FLAG
              'PRRLSN', #PRE-RELIEF REFINANCE LOAN SEQUENCE NUMBER
              'PI', #Program indicator
              'RRI', #RELIEF REFINANCE INDICATOR
              'PVM', #PROPERTY VALUATION METHOD
              'IOI', #INTEREST ONLY INDICATOR
              'MICI',]#MI CANCELLATION INDICATOR
    ODF_df.to_csv(outputpath,index=False)
    print("finished")

def convert_svcg(filepath,outputpath):
    processedData=insertNull(filepath)
    MPDF_df=pd.DataFrame(processedData)
    MPDF_df.columns = ['Id', #LOAN SEQUENCE NUMBER - Unique identifier assigned to each loan
              'MRP', #MONTHLY REPORTING PERIOD
              'CAU', #CURRENT ACTUAL UPB - The Current Actual UPB reflects the mortgage ending balance as reported by the servicer for the corresponding monthly reporting period.
              'CLDS',#CURRENT LOAN DELINQUENCY STATUS
               'LA', #loan age
               'RM2LM',#REMAINING MONTHS TO LEGAL MATURITY
               'DSD', #DEFECT SETTLEMENT DATE
               'MF',#modification flag
                'ZBC', #zero balance code
                'ZBED',#ZERO BALANCE EFFECTIVE DATE
                'CIR',#current interest rate
                'CDU',#CURRENT DEFERRED UPB
                'DDLPI', #DUE DATE OF LAST PAID INSTALLMENT
                'MIR', #MI RECOVERIES
                'NSP', #net sale proceeds
                'NON-MIR', #NON MI RECOVERIES
                'EXPENSES',
                'LC', #legal costs
                'MPC', #MAINTENANCE AND PRESERVATION COSTS
                'TI', #taxes and insurances
                'ME', #MISCELLANEOUS EXPENSES
                'ALC', #ACTUAL LOSS CALCULATION
                'MC', #MODIFICATION COST
                'SMF', #STEP MODIFICATION FLAG
                'DPP', #DEFERRED PAYMENT PLAN
                'ELTV', #ESTIMATED LOAN TO VALUE
                'ZBRU', #ZERO BALANCE REMOVAL UPB
                'DAI', #DELINQUENT ACCRUED INTEREST
                'DDTD', #DELINQUENCY DUE TO DISASTER
                'BASC', #BORROWER ASSISTANCE STATUS CODE
                'CMMC', #CURRENT MONTH MODIFICATION COST
                'IBU',]#INTEREST BEARING UPB
    MPDF_df.to_csv(outputpath,index=False)
    print("finished")
```
